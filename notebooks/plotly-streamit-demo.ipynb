```python
import pandas as pd

data = {
    "Month": ["Jan", "Feb", "Mar", "Apr", "May", "Jun"] * 3,
    "Region": ["North"] * 6 + ["South"] * 6 + ["East"] * 6,
    "Sales": [
        12000, 13500, 15000, 16000, 17000, 18000,
        10000, 11000, 10500, 11500, 12000, 13000,
        8000, 9000, 9500, 10000, 11000, 12500
    ]
}
```

```python
df = pd.DataFrame(data)
df.head()
```

```python
# Using matplotlib

import matplotlib.pyplot as plt

for region in df["Region"].unique():
    subset = df[df["Region"] == region]
    plt.plot(subset["Month"],subset["Sales"],label=region)

plt.title("Monthly Sales Trend by Region")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.legend()
plt.show()
```

```python
# Using plotly

import plotly.express as px

fig = px.line(df,x="Month",y="Sales",color="Region",markers=True,title="Monthly Sales Trend by Region")
fig.show()
```

```python
#comparing Regional Sales Per Month via Barchart

import plotly.express as px

fig = px.bar(df,x="Month",y="Sales",color="Region",barmode="group",title="Monthly Sales Trend by Region")

fig.show()
```

```python
fig2 =  px.bar(
    df,
    x="Product",
    y=("Revenue"),
    color="Region",
    barmode="group",
    title="Revenue by Product and Region"
)

fig2.show()
```

```python
import pandas as pd

data = {
    "Month": ["Jan","Feb","Mar","Apr","May","Jun"] * 3,
    "Region": ["North"]*6 + ["South"]*6 + ["East"]*6,
    "Product": ["Basic","Pro","Enterprise","Basic","Pro","Enterprise"]*3,
    "Revenue": [12000,18000,25000,14000,20000,28000,
                10000,15000,22000,13000,17000,24000,
                8000,12000,18000,10000,15000,21000],
    "Customers": [120,95,40,130,100,50,
                  110,90,35,120,95,45,
                  90,80,30,100,85,40],
    "Churn_Rate": [5.2,4.8,3.1,5.5,4.6,2.9,
                   6.1,5.4,3.8,6.0,5.2,3.4,
                   7.0,6.2,4.5,6.8,6.0,4.0]
}

df = pd.DataFrame(data)
df.head()
```

```python
avg_churn = (
    df.groupby("Product", as_index=False)["Churn_Rate"]
    .mean()
)

fig4 = px.bar(
    avg_churn,
    x="Product",
    y="Churn_Rate",
    color="Product",
    title="Average Churn Rate by Product",
    text="Churn_Rate"
)

fig4.show()
```
